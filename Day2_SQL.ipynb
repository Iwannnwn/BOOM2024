# <b> OPERATOR RASIONAL DAN LOGIKA </b>
Operator rasional dan logika digunakan untuk membandingkan nilai dan menggabungkan kondisi logika. Berikut adalah beberapa contoh operator rasional dan logika yang umum digunakan:

### <b>Operator Rasional</b>
- `==` : Sama dengan
- `!=` : Tidak sama dengan
- `>` : Lebih besar dari
- `<` : Lebih kecil dari
- `>=` : Lebih besar atau sama dengan
- `<=` : Lebih kecil atau sama dengan

### <b>Operator Logika</b>
- `and` : Mengembalikan True jika kedua kondisi adalah True
- `or` : Mengembalikan True jika salah satu kondisi adalah True
- `not` : Membalikkan nilai boolean

### <b>Operator LIKE</b>
- `LIKE` : Digunakan untuk mencocokkan pola dalam teks

### <b>Operator BETWEEN</b>
- `BETWEEN` : Digunakan untuk memilih nilai dalam rentang tertentu

# <b>PENGURUTAN DATA (ORDER BY)</b>

```python
# Pengurutan data pemain yang mengandung kata hunt berdasarkan alfabet
data = pd.read_sql_query(''' SELECT *
                             FROM Player
                             WHERE player_name LIKE "% Hunt%"
                             ORDER BY player_name DESC
                             ''', com)
data
```

# <b> AGREGASI </b>

# <b>GROUP BY</b>

```python
# Mencari tinggi badan dengan total pemain  > 500
data = pd.read_sql_query(''' SELECT height, COUNT(*) AS total_pemain
                             FROM Player
                             GROUP BY 1
                             HAVING total_pemain > 500
                             ''', com)
data
```

```python
# HAVING untuk filter data setelah GROUP BY
data = pd.read_sql_query(''' SELECT height, COUNT(height) AS total_pemain
                                FROM Player
                                GROUP BY 1
                                HAVING total_pemain > 1000
                                ORDER BY 2 DESC
                                ''', com)
data
```

# <b>UNION</b>
UNION dalam SQL adalah operator yang digunakan untuk menggabungkan hasil dari dua atau lebih pernyataan SELECT menjadi satu hasil gabungan. Setiap pernyataan SELECT dalam UNION harus memiliki jumlah kolom yang sama dalam hasilnya dengan tipe data yang kompatibel. Hasil dari UNION hanya akan menampilkan baris yang unik (tidak ada duplikasi).

<b>Contoh penggunaan:</b>

SELECT kolom1, kolom2 FROM tabel1 <br>
UNION <br>
SELECT kolom1, kolom2 FROM tabel2;

<b>Catatan:</b>
- Untuk menyertakan baris duplikat, gunakan UNION ALL.
- Kolom yang digabungkan harus memiliki tipe data yang sama atau kompatibel.

```python
sql = '''SELECT *, 0 AS is_under
         FROM Player
         Where height >= (SELECT AVG(height) FROM Player)
         UNION ALL
         SeLECT *, 1 AS is_under
         FROM Player
         WHERE height < (SELECT AVG(height) FROM Player)
         '''

data = pd.read_sql_query(sql, com)
data
```
