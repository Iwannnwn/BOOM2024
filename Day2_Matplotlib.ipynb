# <b> DATA VISUALITATION WITH MATPLOTLIB </b>
Matplotlib adalah salah satu pustaka (library) paling populer di Python untuk membuat visualisasi data. Dengan Matplotlib, Anda dapat membuat berbagai jenis grafik, seperti grafik garis, grafik batang, histogram, scatter plot, dan banyak lagi. Berikut adalah beberapa konsep dasar dan contoh penggunaan Matplotlib:

## <b>Konsep Dasar</b>
<b>Figure dan Axes:</b>
1. Figure adalah wadah utama untuk semua elemen plot.
2. Axes adalah area di dalam Figure di mana data sebenarnya diplot. Satu Figure bisa memiliki beberapa Axes.

<b>Plotting:</b>
Anda dapat membuat plot dengan menggunakan fungsi seperti plot(), scatter(), bar(), dll.

<b>Customization:</b>
Matplotlib memungkinkan Anda untuk menyesuaikan hampir setiap aspek dari plot, termasuk judul, label sumbu, legenda, warna, dan gaya garis.


```python
# Import Library
import matplotlib.pyplot as plt #type: ignore
import pandas as pd #type: ignore
```

```python
# Import Database
path = 'C:/Users/Asus/OneDrive - Institut Teknologi Sepuluh Nopember/Desktop/DOCS/BOOM 2024/Data/'

df = pd.read_csv(path + 'retail_sales_dataset.csv')
df
```

# <b> GRAFIK GARIS </b>

```python
# Data
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200]

# Plot
plt.plot(x, y, color='blue', marker='o') # marker adalah untuk menandai titik data
plt.title("CONTOH GRAFIK GARIS PENJUALAN")
plt.xlabel("Bulan")
plt.ylabel("Total Penjualan")
plt.show()
```

```python
# Bagaimana untuk dataset df
data = df[df["Date"] <= '2023-03-01']

data.plot(x="Date", y="Total Amount", kind="line")
plt.show()
```

## <b> GRAFIK BATANG </b>

Grafik batang biasanya digunakan untuk membandingkan nilai kategori

```python
# Data
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200]

# Plot Diagram Batang
plt.bar(x, y, color='blue')
plt.title("CONTOH GRAFIK BATANG PENJUALAN")
plt.xlabel("Bulan")
plt.ylabel("Total Penjualan")
plt.show()
```

```python
# Plot Diagram Batang Horizontal
plt.barh(x, y, color='blue')
plt.title("CONTOH GRAFIK BATANG PENJUALAN")
plt.xlabel("Bulan")
plt.ylabel("Total Penjualan")
plt.show()
```

# <b> HISTOGRAM </b>

Histogram digunakan untuk menunjukan distribusi data numerik. Histogram memerlukan data numerik yang akan dikelompokan ke dalam kategori

```python
# Data Histogran
plt.hist(y, bins=5, color='blue', edgecolor='black')
plt.title("CONTOH GRAFIK HISTOGRAM PENJUALAN")
plt.xlabel("Total Penjualan")
plt.ylabel("Frekuensi")
plt.show()
```

# <b> PIE CHART </b>

```python
# Data 
label = ['Category A','Category B', 'Categpry C', 'Category D']
sizes = [25, 35, 25, 15]

# Plot Diagram Pie
plt.pie(sizes, labels=label, autopct='%1.1f%%', startangle=90)
plt.title("CONTOH GRAFIK PIE PENJUALAN")
plt.show()
```

# <b>SCATTER PLOT</b>
Scatter plot adalah jenis grafik yang digunakan untuk menampilkan hubungan antara dua variabel numerik. Setiap titik pada scatter plot mewakili satu pengamatan dalam dataset. Scatter plot sangat berguna untuk mengidentifikasi pola, tren, dan korelasi antara dua variabel.

Beberapa ketentuan dalam membuat scatter plot:
1. **Sumbu X dan Y:** Scatter plot membutuhkan dua variabel numerik, satu untuk sumbu X dan satu untuk sumbu Y.
2. **Titik Data:** Setiap titik pada scatter plot mewakili satu pengamatan dalam dataset.
3. **Korelasi:** Scatter plot dapat digunakan untuk mengidentifikasi korelasi antara dua variabel. Korelasi positif ditunjukkan oleh pola titik yang naik dari kiri bawah ke kanan atas, sedangkan korelasi negatif ditunjukkan oleh pola titik yang turun dari kiri atas ke kanan bawah.
4. **Outliers:** Scatter plot juga dapat membantu mengidentifikasi outliers atau pengamatan yang jauh dari pola umum data.

```python
# Data
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]

# Plot Diagram Scatter
plt.scatter(x, y, color='blue', marker='o')
plt.title("CONTOH GRAFIK SCATTER")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.show()
```

## <b> Menambahkan Grid </b>

```python
# Grafik dengan Grid
plt.plot(x,y)
plt.grid(True)
plt.title("Grafik Dengan Grid")
plt.show()
```

## <b> Menambahkan Beberapa Grafik dalam Satu Gambar </b>

```python
# Data
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,3,6,10,15]

# Membuat Grafik
plt.plot(x, y1, label="y = x^2", color='blue')
plt.plot(x, y2, label="y = x(x+1)/2", color='red')
plt.title("Grafik Dua Fungsi")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() # Menampilkan Legenda
plt.show()
```

## <b>Mengubah Ukuran dan Resolusi </b>

```python
plt.figure(figsize=(10,5), dpi=100)
plt.plot(x, y1)
plt.title("Ukuran dan Resolusi Grafik")
plt.show()
```

# <b>SUBPLOT</b>

```python
# Data
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,3,6,10,15]

# Membuat subplot 1 (1 baris, 2 kolom, grafik ke-1)
plt.subplot(1,2,1)
plt.plot(x, y1, color='blue')
plt.title("y=x^2")

# Membuat subplot 2 (1 baris, 2 kolom, grafik ke-2)
plt.subplot(1,2,2)
plt.plot(x, y2, color='red')
plt.title("y=x(x+1)/2")

plt.suptitle("Contoh Subplot")
plt.show()
```
